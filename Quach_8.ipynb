```python
import pandas as pd
```

```python
import seaborn as sns
```

```python
df = pd.read_csv('avocado.csv')
df.head(5)
```

```python
df.info()
```

```python
df['Date'] = pd.to_datetime(df['Date'])
df.info()
```

```python
df.query('year == "2015" and type != "conventional" and region != "Albany"')
bg = df[['Date', 'Total Bags', 'Small Bags']]
bg.head(5)
```

```python
bag = df.melt(id_vars ='Date', value_vars = ['Total Bags', 'Small Bags'], var_name = 'Bags', value_name = 'Count')
display(bag)
```

```python
sns.relplot(data = bag, kind = 'line', x = 'Date', y = 'Count', hue = 'Bags')
```

```python
df['Sales Volume'] = pd.qcut(df['Total Volume'], q=4, labels = ['Poor', 'Modest', 'Good', 'Excellent'])
df.head(5)
```

```python
sns.countplot(data = df, x = 'year', hue = 'Sales Volume')
```

```python
bag = df.melt(id_vars ='year', value_vars = ['Small Bags', 'Large Bags', 'XLarge Bags'], var_name = 'Bags', value_name = 'Count')
display(bag)
```

```python
sns.relplot(data = bag, kind = 'line', x = 'year', y = 'Count', hue = 'Bags')
```
